````python
# @title
# define add_total_column_and_reshape_data_sql function to:
# - convert clean dataframe to a reshaped SQL results table with 2 columns - 'State'
#   & 'Total State Consumption (PJ)' (sum of all of state's various energy consumption types);
# - convert resulting results table into dataframe suitable for pie chart
def add_total_column_and_reshape_data_sql():
  '''
  Add a total consumption onto a clean dataframe from the clean SQL table,
  and reshape dataframe narrowing towards total energy consumption.
  Assumes connection to clean db file has already been made.
  '''
  #create total column; only show states and total consumption
  sql='''
  SELECT
      State,
      SUM("Coal (PJ)" + "Oil (PJ)" + "Gas (PJ)" + "Renewables (PJ)") AS 'Total State Consumption (PJ)'
  FROM
      AUS_energy_clean
  GROUP BY
      State;
  '''

  cursor.execute(sql)
  cursor.fetchall()

  # create dataframe from sql query
  total_cons_state_pj_df = pd.read_sql_query(sql, conn)

  # return dataframe
  return total_cons_state_pj_df

# define function to create pie chart
# with whole pie representing total Australian energy consumption for
# the whole period contained in the database,
# and 1 slice representing 1 state's energy consumption for the whole period
def plot_pie_chart_nat_energy_from_clean_reshaped_df(total_cons_state_pj_df):
  '''
  Create pie chart, where whole pie represents total Australian energy
  consumption for entire historical period contained in the data. Each slice
  represents 1 state/territory's energy consumption for that period.
  '''

  # turn data in dataframe into the appropriate lists needed for pie chart plot
  states_labels = total_cons_state_pj_df['State']
  states_total_cons = total_cons_state_pj_df['Total State Consumption (PJ)']

  # create setting for pie chart, set size of piechart
  plt.figure(figsize=(8, 6))
  # draw actual piechart
  plt.pie(states_total_cons, labels=states_labels, autopct='%1.1f%%', startangle=90)
  # create title for pie 'Total Energy Consumption by State from 2008-2022 (PJ)'
  plt.title('Total Energy Consumption by State from 2008-2022')
  plt.axis('equal')
  # show chart in setting with title
  plt.show()

# define function to take clean version of Aus_energy_sqlite3.db database,
# and plot pie chart showing total historical consumption for Australia for whole period,
# broken down into total consumption of each state for whole period
def plot_pie_chart_nat_energy_from_clean_unshaped_dataframe(df_clean):
  # map clean dataframe to SQL table
  map_clean_df_to_sql_table(df_clean)

  # call add_total_column_and_reshape_data_sql() function to perform necessary
  # calculations on clean SQL table and create dataframe in shape necessary
  # for pie chart plotting
  total_cons_state_pj_df = add_total_column_and_reshape_data_sql()

  # plot appropriate pie chart
  plot_pie_chart_nat_energy_from_clean_reshaped_df(total_cons_state_pj_df)

# define all_states_cons_brkdwn_clean_sql_to_reshaped_df function to draw on
# existing SQL table with clean data (made earlier) to:
# - convert clean dataframe to a reshaped SQL results table with 5 columns - 'State'
#   and 4 columns for each of the various energy consumption types;
# - convert resulting results table into dataframe suitable for pie chart
def all_states_cons_brkdwn_clean_sql_to_reshaped_df():
  '''
  Select 'State', 'Coal (PJ)', 'Oil (PJ)' , 'Gas (PJ)', 'Renewables (PJ)' from
  clean SQL table, and map this SQL results table onto a dataframe using pandas.
  Assumes connection to clean db file has already been made.
  '''

  # from clean SQL table, creates results table which only shows
  # column for states and columns for each different type of energy consumption
  sql='''
  SELECT
      State, "Coal (PJ)" , "Oil (PJ)" , "Gas (PJ)" , "Renewables (PJ)"
  FROM
      AUS_energy_clean
  GROUP BY
      State;
  '''
  cursor.execute(sql)
  cursor.fetchall()

  # create df from results table with states and energy-consumption-type columns
  all_states_cons_brkdwn_df = pd.read_sql_query(sql, conn)

  # return dataframe
  return all_states_cons_brkdwn_df

  # define function to create pie chart with whole pie representing total energy
# consumption for 1 state for the whole period contained in the database, and
# 1 slice representing 1 type of energy consumption for that state for
# the whole period
def plot_pie_chart_state_cons_brkdwn_frm_clean_reshaped_df(chosen_state,all_states_cons_brkdwn_df):
  '''
  Create pie chart, where whole pie represents total energy consumption for 1
  state for entire historical period contained in the data. Each slice
  represents 1 type of energy consumption for that state for that period.
  '''

  # filter dataframe to get data for chosen state only
  chosen_state_cons_brkdwn_df = all_states_cons_brkdwn_df.loc[all_states_cons_brkdwn_df['State'] == chosen_state]
  # locate rows for chosen state's data, for each type of energy consumption
  coal_cons = chosen_state_cons_brkdwn_df.iloc[0]["Coal (PJ)"]
  oil_cons = chosen_state_cons_brkdwn_df.iloc[0]["Oil (PJ)"]
  gas_cons = chosen_state_cons_brkdwn_df.iloc[0]["Gas (PJ)"]
  renewables_cons = chosen_state_cons_brkdwn_df.iloc[0]["Renewables (PJ)"]

  # turn dataframe data into appropriate lists needed for pie chart plot
  cons_type_labels = ['Coal', 'Oil', 'Gas', 'Renewables']
  cons_type_amount = [coal_cons, oil_cons, gas_cons, renewables_cons]

  # create setting for pie chart, set size of piechart
  plt.figure(figsize=(8, 6))
  # draw actual piechart
  plt.pie(cons_type_amount, autopct='%1.1f%%', startangle=90)
  # create title and legend for pie chart
  plt.title(f'Total Energy Consumption of {chosen_state} from 2008-2022')
  plt.legend(cons_type_labels, bbox_to_anchor=(1,1))
  plt.axis('equal')
  #show chart in setting with title and legend
  plt.show()

# define function to take clean version of Aus_energy_sqlite3.db database,
# and plot pie chart showing total historical consumption for single state for
# whole period contained in dataset, broken down into energy sources
def plot_pie_chart_chosen_state_cons_brkdwn_from_clean_unshaped_dataframe(chosen_state,df_clean):
  # map clean dataframe to SQL table
  map_clean_df_to_sql_table(df_clean)

  # call all_states_cons_brkdwn_clean_sql_to_reshaped_df()() function to select
  # only state & energy breakdown columns from clean SQL table,
  # and create dataframe in shape necessary for pie chart plotting
  all_states_cons_brkdwn_df = all_states_cons_brkdwn_clean_sql_to_reshaped_df()

  # plot appropriate pie chart
  plot_pie_chart_state_cons_brkdwn_frm_clean_reshaped_df(chosen_state,all_states_cons_brkdwn_df)

# define function to ask user their choice (nat/state) and call appropriate
# function to make that plot
def da1_nat_or_chosen_state_analysis(df_clean, choice, chosen_state):
  '''
  Plot 2008-2022 energy consumption for: whole country OR 1 state, based on
  user choice. Assumes clean dataset already connected.
  '''

  if choice == 'state':
    # call function to reshape data and plot pie for user's chosen state
    plot_pie_chart_chosen_state_cons_brkdwn_from_clean_unshaped_dataframe(chosen_state,df_clean)

  elif choice == 'national':
    # call function to plot pie for whole of Australia
    plot_pie_chart_nat_energy_from_clean_unshaped_dataframe(df_clean)

  else:
    print("Only analyse in a state-level or national-scale.")
````

````python
# @title
# define add_total_column_state_year_sql_to_df_reshape_for_cc function to:
# - convert clean dataframe to a reshaped SQL results table with 2 columns - 'State'
#   & 'Total State Consumption (PJ)' (sum of all of state's various energy
#   consumption types, grouped by state & year);
# - convert resulting results table into dataframe
# - reshape df so that it is suitable for a clustered column chart
def add_total_column_state_year_sql_to_df_reshape_for_cc():
  '''
  From clean dataframe, reshape data (by converting to SQL table & back) so that
  it has appropriate structure for clustered column chart.
  Assumes connection to clean db file has already been made.
  '''
  # Call out tables and group them by state then year to visualise how the data
  # will look in a cluster diagram
  sql='''
  SELECT
      Year, State,
      SUM("Coal (PJ)" + "Oil (PJ)" + "Gas (PJ)" + "Renewables (PJ)") AS 'total_state_consumption (PJ)'
  FROM
      AUS_energy_clean
  GROUP BY
      State, Year;
  '''
  cursor.execute(sql)
  cursor.fetchall()

  # map makes df
  total_cons_state_per_year_pj_df = pd.read_sql_query(sql, conn)
  total_cons_state_per_year_pj_df

  # Group the data by State and Year -> has effect of ordering it
  grouped_energy = total_cons_state_per_year_pj_df.groupby(['State', 'Year'])
  # Show the values as sum of each value
  grouped_energy_values_df = grouped_energy.sum()

  # Reshape the result to have State as rows and Year as columns
  grouped_energy_values_df = grouped_energy_values_df.unstack()

  # return resulting DataFrame
  return grouped_energy_values_df

# define function to create clustered column chart with each column cluster
# representing one state over the 14 years (2008-2022), and each individual
# column within a single cluster representing total energy consumption (in PJ)
# for that state within a single year.
def plot_cc_chart_all_states_cons_clusters_frm_clean_reshaped_df(grouped_energy_values_df):
  '''
  Create a clustered column chart for clean reshaped dataset, with each column
  cluster representing one state over the 14 years (2008-2022), and each
  individual column within a single cluster representing total energy
  consumption (in PJ) for that state within a single year.
  '''
  # create list of contrasting colours to be used for each column per cluster
  colors = ['blue', 'orange', 'limegreen', 'red', 'purple', 'brown', 'darkgreen', 'magenta', 'olivedrab', 'darkturquoise', 'darkslateblue', 'gold', 'orchid', 'maroon']
  # draw actual clustered columns in those colours
  ax = grouped_energy_values_df.plot(kind='bar', figsize=(10, 5), color=colors)
  # create title, axis labels, and legend
  plt.title('Annual Energy Consumption Distribution by State and Year', fontsize = '18')
  plt.xlabel('State')
  plt.ylabel('Annual Energy Consumption (PJ)')
  plt.legend(title='Alignment', loc='upper right', bbox_to_anchor=(1.20, 0.5), labels=['2008-09', '2009-10', '2010-11', '2011-12', '2012-13', '2013-14', '2014-15', '2016-17', '2018-19', '2019-20', '2020-21', '2021-22'])
  # rotate 'Year' ticks along x-axis
  plt.xticks(rotation=0)

  #show chart with title and legend
  plt.show()

# define reshape_total_cons_df_for_area_chart_state_energy function to filter and
# reshape data from df_clean dataframe, to prepare it for shape required by
# area chart reflecting composition of individual state's energy consumption
def reshape_total_cons_df_for_area_chart_state_energy(dataframe,state):
  '''
  Takes input 'dataframe', removes 'State', 'Population' and 'GSP ($ million)'
  columns, and filters by input 'state'. As long as 'dataframe' input has
  'Year', 'State', 'Coal (PJ)', 'Oil (PJ)', 'Gas (PJ)', 'Renewables (PJ)',
  'Population', and 'GSP ($ million)' columns, function will output dataframe
  with 'Year', 'State', 'Coal (PJ)',	'Oil (PJ)', 'Gas (PJ)' and
  'Renewables (PJ)' columns, filtered by input 'state'.
  This output dataframe can be used to create an area chart reflecting
  composition of an individual state's energy consumption over time.
  '''

  # copy input dataframe so modifications made to new dataframe
  # do not affect input dataframe
  df_clean_chosen_state = dataframe.copy()

  # filter by state
  df_clean_chosen_state = df_clean_chosen_state[df_clean_chosen_state['State'] == state]

  # create list of names of columns to drop
  state_area_columns_to_drop = ['State','Population','GSP ($ million)']

  # drop previously identified columns
  df_clean_chosen_state_area = df_clean_chosen_state.drop(columns=state_area_columns_to_drop)

  # set Year as index
  df_clean_chosen_state_area = df_clean_chosen_state_area.set_index('Year')

  # return dataframe
  return df_clean_chosen_state_area

# define function to take clean version of Aus_energy_sqlite3.db database,
# and plot clusterd column chart
def plot_cc_chart_all_states_cons_clusters_frm_clean_df(df_clean):
  '''
  Plot clustered column chart of total annual energy consumption for each over
  time, with each cluster of columns representing one state and individual
  columns representing consecutive years of total annual energy consumption for
  that state.
  '''
  # map clean dataframe to SQL table
  map_clean_df_to_sql_table(df_clean)

  # call add_total_column_state_year_sql_to_df_reshape_for_cc function to
  # reshape data as necessary for clustered column chart, store resulting df
  # into variable 'grouped_energy_values_df'
  grouped_energy_values_df = add_total_column_state_year_sql_to_df_reshape_for_cc()

  # plot appropriate clustered column chart using reshaped df
  plot_cc_chart_all_states_cons_clusters_frm_clean_reshaped_df(grouped_energy_values_df)

# define reshape_total_cons_df_for_area_chart_state_energy function to filter and
# reshape data from df_clean dataframe, to prepare it for shape required by
# area chart reflecting composition of individual state's energy consumption
def reshape_total_cons_df_for_area_chart_state_energy(df_clean,chosen_state):
  '''
  Takes input 'dataframe', removes 'State', 'Population' and 'GSP ($ million)'
  columns, and filters by input 'state'. As long as 'dataframe' input has
  'Year', 'State', 'Coal (PJ)', 'Oil (PJ)', 'Gas (PJ)', 'Renewables (PJ)',
  'Population', and 'GSP ($ million)' columns, function will output dataframe
  with 'Year', 'State', 'Coal (PJ)',	'Oil (PJ)', 'Gas (PJ)' and
  'Renewables (PJ)' columns, filtered by input 'state'.
  This output dataframe can be used to create an area chart reflecting
  composition of an individual state's energy consumption over time.
  '''

  # copy input dataframe so modifications made to new dataframe
  # do not affect input dataframe
  df_clean_chosen_state = df_clean.copy()

  # filter by state
  df_clean_chosen_state = df_clean_chosen_state[df_clean_chosen_state['State'] == chosen_state]

  # create list of names of columns to drop
  state_area_columns_to_drop = ['State','Population','GSP ($ million)']

  # drop previously identified columns
  df_clean_chosen_state_area_reshaped = df_clean_chosen_state.drop(columns=state_area_columns_to_drop)

  # set Year as index
  df_clean_chosen_state_area_reshaped = df_clean_chosen_state_area_reshaped.set_index('Year')

  # return dataframe
  return df_clean_chosen_state_area_reshaped, chosen_state

# define function to create area plot with breakdown of total energy consumption,
# by energy type, for single state, over time
def plot_area_chart_state_energy(df_clean_chosen_state_area_reshaped, chosen_state):
  '''
  Create area chart for a clean reshaped dataset, showing breakdown of
  energy consumption by type, for a single input 'state', over time.
  '''

  # set title using f-string to insert chosen state name in it
  plot_title = f'Breakdown of {chosen_state} Energy Consumption, 2008-2022'
  y_axis_label = 'Energy Consumption (PJ)'

  # plot using area chart option (which is stacked by default)
  df_clean_chosen_state_area_reshaped.plot.area(stacked=True,title=plot_title,ylabel=y_axis_label,rot=45)

# define function to take clean version of Aus_energy_sqlite3.db database,
# and plot clusterd column chart
def plot_area_chart_chosen_state_frm_clean_df(df_clean,chosen_state):
  # call function to reshape clean dataset for area chart, feed returned df into variable
  df_clean_chosen_state_area_reshaped, chosen_state = reshape_total_cons_df_for_area_chart_state_energy(df_clean,chosen_state)

  # plot appropriate clustered column chart using reshaped df
  plot_area_chart_state_energy(df_clean_chosen_state_area_reshaped, chosen_state)

# define function to ask user their choice (nat/state) and call appropriate
# function to make that plot
def da2_nat_or_chosen_state_analysis(df_clean, choice, chosen_state):
  '''
  Plot annual energy consumption for: whole country OR 1 state, based on
  user choice. Assumes clean dataset already connected.
  '''

  if choice == 'state':
    # call function to reshape data and plot area chart for user's chosen state
    plot_area_chart_chosen_state_frm_clean_df(df_clean,chosen_state)

  elif choice == 'national':
    # call function to plot clustered column chart for whole of Australia
    plot_cc_chart_all_states_cons_clusters_frm_clean_df(df_clean)

  else:
    print("Only analyse in a state-level or national-scale.")
````

### **Data Analysis 3 and 4: Exploring the Annual Energy Consumption of States in Australia per Capita and Unit of GSP $million**

**Note:** Before starting we need to define various functions that will help hasten the process of the data analysis process. The following are the functions: *Refer to Development notebook for further clarification.*
* ***convert_pj_to_gj(pj_value):*** Converts PJ values to GJ values.
  * Returns *gj_value*


---


* ***convert_gj_to_pj(gj_value):*** Converts GJ values to PJ values.
  * Returns *pj_value*

* ***add_total_column(dataframe):*** Adds a new column for total energy consumption (PJ) into a copy of the clean dataframe created within the function.
  * Returns *df_clean_w_total_cons*


---


* ***add_energy_cons_per_cap(dataframe):*** Adds a new column for total energy consumption per capita (GJ/person) into a copy of the dataframe with total consumption created within the function.
  * Returns *df_clean_w_cons_per_cap*


---


* ***add_energy_cons_per_gsp(dataframe):*** Adds a new column for total energy consumption per GSP million (GJ/GSP $million) into a copy of the dataframe with total consumption per capita created within the function.
  * Returns *df_clean_w_cons_per_gsp*

Dataframe is pivoted with only the values *State*, *Year*, and *Consumption Per Capita (GJ/person)*.
  * Returns *df_clean_w_cons_per_cap_reshaped*

* ***reshape_cons_per_gsp_df_for_multiline_plot(dataframe):*** Similar function to per capita reshaping, but Dataframe is pivoted with only the values *State*, *Year*, and *Consumption Per GSP (GJ/GSP million)*.
  * Returns *df_clean_w_cons_per_gsp_reshaped*


---


* ***plot_multiline_cons_per_cap_all_states(dataframe):*** Plots the reshaped df for total consumption per capita into a multiline chart.

````python
# @title
# create function to convert GJ to PJ & vice versa
# given that 1 PJ = 1,000 GJ
  # create function to convert GJ to PJ
def convert_pj_to_gj(pj_value):
  '''Convert PJ to GJ (*1000)'''
  gj_value = pj_value*1000
  return gj_value

  # create function to convert GJ to PJ
def convert_gj_to_pj(gj_value):
  '''Convert GJ to PJ (/1000)'''
  pj_value = gj_value/1000
  return pj_value


# create function that will add a new column for total consumption into a copy of the clean dataframe
  # using pre-existing dataframe returned as df_clean_w_total_cons
def add_total_column(dataframe):
  '''
  Add 'Total Energy (PJ)' column (calculated as the sum of
  'Coal (PJ)','Oil (PJ)','Gas (PJ)' and 'Renewables (PJ)' for that row)
  to an existing dataframe.
  '''
  # create duplicate of clean df,
  # which we can work with without affecting clean version of original df
  df_clean_w_total_cons = df_clean.copy()

  # create list with names of all columns relating to
  # different types of energy consumption
  energy_column_names = ['Coal (PJ)','Oil (PJ)','Gas (PJ)','Renewables (PJ)']

  # create list with all columns relating to different
  # types of energy consumption, contextualised against
  # their location in the df_clean_w_total_cons dataframe
  # ie. [dataframe['energy_column_name1'],dataframe['energy_column_name2']]
  # rather than just ['energy_column_name1','energy_column_name2'].
  # This arrangement was necessary because the sum() function required
  # the format '[dataframe['energy_column_name1'],dataframe['energy_column_name2']]'
  # rather than just ['energy_column_name1','energy_column_name2'].
  energy_column_names_incontext = []
  for name in energy_column_names:
    energy_column_names_incontext.append(df_clean_w_total_cons[name])

  # ADD column with total energy
  df_clean_w_total_cons['Total Energy (PJ)']=sum(energy_column_names_incontext)

  # check column added
  df_clean_w_total_cons.head(2)

  # return resulting dataframe
  return df_clean_w_total_cons


# define function to add 'Consumption Per Capita (GJ/person)' column to
  # using pre-existing dataframe returned as df_clean_w_cons_per_cap
def add_energy_cons_per_cap(dataframe):
  '''
  Add 'Consumption Per Capita (GJ/person)' column to an exsting dataframe.
  'Consumption Per Capita (GJ/person)' calculated as sum of 'Coal (PJ)',
  'Oil (PJ)','Gas (PJ)' and 'Renewables (PJ)' energy types,
  divided by the population of that state,
  for the year and state to which that row relates.
  '''
  # call function to add 'Total Energy (PJ)' column to cleaned dataframe
  df_clean_w_total_cons = add_total_column(dataframe)

  # create duplicate df to work with for 'Consumption Per Capita' analysis
  df_clean_w_cons_per_cap = df_clean_w_total_cons.copy()

  # establish 'population' and 'totalEnergy' variables,
  # in relation to the df_clean_w_cons_per_cap dataframe
  population = df_clean_w_cons_per_cap['Population']
  total_energy = df_clean_w_cons_per_cap['Total Energy (PJ)']

  # add calculated 'Consumption Per Capita (PJ)' column
  df_clean_w_cons_per_cap['Consumption Per Capita (GJ/person)'] = convert_pj_to_gj(total_energy/population)

  # return the newly created dataframe
  return df_clean_w_cons_per_cap


# define function to add 'Consumption Per Unit of GSP (GJ/$ million)' column to
  # using pre-existing dataframe returned as df_clean_w_cons_per_gsp
def add_energy_cons_per_gsp(dataframe):
  '''
  Add 'Consumption Per Unit of GSP (GJ/$ million)' column to an exsting dataframe.
  'Consumption Per Unit of GSP (GJ/$ million)' calculated as sum of 'Coal (PJ)',
  'Oil (PJ)','Gas (PJ)' and 'Renewables (PJ)' energy types,
  divided by the GSP for the year and state to which that row relates.
  '''
  # call function to add 'Total Energy (PJ)' column to cleaned dataframe
  df_clean_w_total_cons = add_total_column(dataframe)

  # create duplicate df to work with for 'Consumption Per Unit of GSP (GJ/$ million)' analysis
  df_clean_w_cons_per_gsp = df_clean_w_total_cons.copy()

  # establish 'gsp' and 'total_energy' variables,
  # in relation to the df_clean_w_cons_per_gsp dataframe
  gsp = df_clean_w_cons_per_gsp['GSP ($ million)']
  total_energy = df_clean_w_cons_per_gsp['Total Energy (PJ)']

  # add calculated 'Consumption Per Unit of GSP (GJ/$ million)' column
  df_clean_w_cons_per_gsp['Consumption Per Unit of GSP (GJ/$ million)'] = convert_pj_to_gj(total_energy/gsp)

  # return the newly created dataframe
  return df_clean_w_cons_per_gsp


# define reshape_cons_per_cap_df_for_multiline_plot function to reshape data from
# df_clean_w_cons_per_cap dataframe, to prepare it for shape required by multi-line plot
def reshape_cons_per_cap_df_for_multiline_plot(dataframe):
  '''
  Reshape dataframe with 'Year', 'State' and 'Consumption Per Capita (GJ/person)'
  columns, into dataframe with States as columns, years as rows, and
  Consumption Per Capita (GJ/person) as values.
  '''
  # call add_energy_cons_per_cap with input dataframe,
  # in order to do calculations for & add columns for 'Consumption Per Capita' metric
  # & save returned dataframe as 'df_clean_w_cons_per_cap'
  df_clean_w_cons_per_cap = add_energy_cons_per_cap(dataframe)

  # from df_clean_w_cons_per_cap dataframe, create new reshaped dataframe, 'df_clean_w_cons_per_cap_reshaped', using pandas 'pivot' method
  df_clean_w_cons_per_cap_reshaped = df_clean_w_cons_per_cap.pivot(index='Year',columns='State',values='Consumption Per Capita (GJ/person)')

  # return newly created, reshaped, dataframe
  return df_clean_w_cons_per_cap_reshaped


# define reshape_cons_per_gsp_df_for_multiline_plot function to reshape data from
# df_clean_w_cons_per_gsp dataframe, to prepare it for shape required by multi-line plot
def reshape_cons_per_gsp_df_for_multiline_plot(dataframe):
  '''
  Reshape dataframe with 'Year', 'State' and 'Consumption Per Unit of GSP (GJ/$ million)'
  columns, into dataframe with States as columns, years as rows, and
  Consumption Per Unit of GSP (GJ/$ million) as values.
  '''
  # call add_energy_cons_per_gsp with input dataframe,
  # in order to do calculations for & add columns for 'Consumption Per Unit of GSP' metric
  # & save returned dataframe as 'df_clean_w_cons_per_gsp'
  df_clean_w_cons_per_gsp = add_energy_cons_per_gsp(dataframe)

  # from df_clean_w_cons_per_gsp dataframe, create new reshaped dataframe, 'df_clean_w_cons_per_gsp_reshaped', using pandas 'pivot' method
  df_clean_w_cons_per_gsp_reshaped = df_clean_w_cons_per_gsp.pivot(index='Year',columns='State',values='Consumption Per Unit of GSP (GJ/$ million)')

  # return newly created, reshaped, dataframe
  return df_clean_w_cons_per_gsp_reshaped


# define function to create a multiline graph showing changes in
# energy consumption per capita for all states, over time.
def plot_multiline_cons_per_cap_all_states(dataframe):
  '''
  Create multiline graph for a single clean dataset, 'dataframe', showing changes in
  total energy consumption per capita for all Australian states over time.
  '''
  # create dataframe with added columns/calculations, in necessary shape for multiline plot
  df_clean_w_cons_per_cap_reshaped = reshape_cons_per_cap_df_for_multiline_plot(dataframe)

  # store title of plot in variable 'plot_title'
  plot_title='''Energy Consumption Per Capita
   Per Australian State/Territory, 2008-2022'''
  # store label for y-axis in variable, 'y_axis_label'
  y_axis_label="Energy Consumption Per Capita (GJ/person)"

  # plot body of multiline plot for Energy Consumption Per Capita Per State over time,
  # with title as 'plot_title' variable and y-axis label as 'y_axis_label' variable
  df_clean_w_cons_per_cap_reshaped.plot(title=plot_title,ylabel=y_axis_label,rot=45)

  # set location of legend to upper left
  plt.legend(loc='upper left',fontsize=9)

  # show plot
  plt.show()


def plot_multiline_cons_per_gsp_all_states(dataframe):
  '''
  Create multiline graph for a single clean dataset, 'dataframe', showing changes in
  total energy consumption per unit of GSP for all Australian states over time.
  '''
  # create dataframe with added columns/calculations, in necessary shape for multiline plot
  df_clean_w_cons_per_gsp_reshaped = reshape_cons_per_gsp_df_for_multiline_plot(dataframe)

  # store title of plot in variable 'plot_title'
  plot_title='''Energy Consumption Per Unit of GSP
   Per Australian State/Territory, 2008-2022'''
  # store label for y-axis in variable, 'y_axis_label'
  y_axis_label="Energy Consumption Per Unit of GSP (GJ/$ million)"

  # plot body of multiline plot for Energy Consumption Per Unit of GSP Per State over time,
  # with title as 'plot_title' variable and y-axis label as 'y_axis_label' variable
  df_clean_w_cons_per_gsp_reshaped.plot(title=plot_title,ylabel=y_axis_label,rot=45)

  # set location of legend to upper left
  plt.legend(loc='upper left',fontsize=8)

  # show plot
  plt.show()
````

### Why is it a good practice? (Explain why closing connection is highly recommended by the end of the session)

In the context of this data, closing the DB will not be a compulsory practice. However, when working with data in general, closing the DB is a *COMPULSORY PRACTICE* because of the following reasons: (hnurn, 2024)

* **Security:** Leaving the database continuously opened without closing is like leaving the doors unlocked after leaving a house. Regardless how limited the access is, if it is out in the open, any highly-skilled individuals can easily alter/steal information stored in the database.
* **Traffic:** Try driving in an intersection with all lanes having green lights non-stop. Either it is being eager for an accident to happen or traffic to come and halt the entire flow.
* **Resource management:** When the database is still running in the background, this will influence the input and output process. Without closing, any new data/information added will go to that particular database without the user's knowledge, altering the existing database in the process.
* **Performance:** Having the database to continually be open also means having the computer constantly calculate information that is also inside the database. As a result, a portion of the available RAM is taken away, reducing the computer's performance.
